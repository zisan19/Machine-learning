While Loop

```python
count=1
while count<10:
    print(count,"Hello")
    count+=1
print(count)
```

```python
sum=0
for i in range(1,11):
    print(i,"Is processing")
    if i%2==0: ##even
        continue
    
    sum+=i
    print(i,"Is added to the sum")
    
print(sum)
```

```python
accuracy=95

changed_or_not=False
while True:
    if changed_or_not==False and accuracy==100:
        break
    
    # if(prev>accuracy):
    #     changed_or_not=False
    accuracy+=1
    print(accuracy)
```

```python
t=int(input())
for i in range(t):
    num=int(input())
    if num==0:
        print(0)
        continue
    
    while num>0:
        print(num%10,end=" ")# num=int(num/10)
        num//=10
    print()
```
