```python
a_friends = {"Rahim", "Karim", "Sakib", "Jamal"}
b_friends = {"Sakib", "Jamal", "Rafiq", "Nadim"}

Mutual=a_friends & b_friends 
Unique_to_a=b_friends-a_friends
Unique_to_b=a_friends-b_friends
Total_unique_friends=a_friends | b_friends

print(f"Mutual friends:",Mutual)
print(f"Unique to A:",Unique_to_a)
print(f"Unique to B:",Unique_to_b)
print(f"Total unique friends:",len(Total_unique_friends))
    
```

```python
students = [("Rafi", 89), ("Sumi", 95), ("Hasan", 90), ("Nila", 75), ("Anik", 98)]
desc=sorted(students,key=lambda i:i[1],reverse=True)
top_3=desc[:3]
print("Top 3 students:")
for students in top_3:
    print(f"{students[0]}-{students[1]}")
```
