```python
from functools import reduce
with open("sample.txt","r") as file:
    strings=file.readlines()
    print(strings)
    
    #total line koyta
    total_lines=len(strings)
    print(total_lines)
    
    #Word ber korbo
    Number_of_Words=list(map(lambda x:len(x.split()),strings))
    
    #Total words
    total_num_of_words=reduce(lambda x,y:x+y,Number_of_Words)
    print(total_num_of_words)
    
```
