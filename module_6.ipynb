```python
def square_addition(a,b,c):
    print(f"{a} {b} {c}")
    a=a**2 
    b= b**2 
    return a+b+c 
#positional argument
ans = square_addition(10,20,30) + 10 
print(ans) 

#keyword argument
ans = square_addition(b=10,c=20,a=30) + 10 
print(ans) 
```

```python
def square_addition(*args):
    sum=0
    for i in args:
        i=i**2 
        sum+=i
    return sum
```

##### Generator

```python
lst=[x for x in range(500)] #dataset

def data_loader(chunk_size,list):
    for i in range(len(lst)):
        yield lst[i:i+chunk_size]
        
x=data_loader(5,lst)
print(type(x))
print(x)
print(next(x))
```

```python
lst=[x for x in range(500)] #dataset

def data_loader(chunk_size,list):
    for i in range(0,len(lst),chunk_size):   #Problem
        yield lst[i:i+chunk_size]
        
x=data_loader(5,lst)
print(type(x))
print(x)
print(next(x))
print(next(x))
```
